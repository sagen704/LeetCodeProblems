## Description
Given an array of intervals where intervals[i] = [starti, endi], merge all overlapping intervals, and return an array of the non-overlapping intervals that cover all the intervals in the input.


## My First Solution

So I sorted the list of arrays and that sorted them by the first number in there index and then from there I just went through and checked if I needed to merge the intervals, or not.

I went through and I was making an interval if I could merge it with the next one. If it completely covered the next one I did not even bother. If it did not spill into the next one I added the cleaned one to the output and then I started looking at the next one.

```python
class Solution:
    def merge(self, intervals: List[List[int]]) -> List[List[int]]:

        # going to go through and sort them and then just loop through once

        output = []

        sorted_interval = sorted(intervals)

        current_interval = sorted_interval[0]

        for num_set in sorted_interval:

            # they are sorted so current interval[0] is the start of the whole list
            # Need to check if you can expand the interval,
            # if not add the current interval to the output
            # if can update it and keep moving forward

            if (current_interval[1] >= num_set[0]):
                # the second one of current interval spills into next interval can expand it

                # Check that the current interval does not cover entire num_set
                if not (current_interval[1] > num_set[1]):
                    current_interval[1] = num_set[1]

            else:
                output.append(current_interval)
                current_interval = num_set

            # print(f'Current interval:{current_interval}')
            # print(f'Num_set:{num_set}')
            # print(output)
            # print()

        output.append(current_interval)
        return output
```
